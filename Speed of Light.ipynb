```python
import matplotlib.pyplot as plt
import numpy as np
```

# First we will start with the preliminary data for 1 Mhz

```python
# Original data
x_data = np.array([17.88, 44.81, 79.76])
y_data = np.array([-0.00000022,-0.000000292,0.000000075])

#Red data points
x_new_data = np.array([17.88, 44.81, 79.76])
y_new_data = np.array([-0.00000022, 0.000000708,0.000002075])

#Polynomial Fitting
coefficients = np.polyfit(x_new_data, y_new_data, 1)
# Create a polynomial object using the coefficients
polynomial = np.poly1d(coefficients)

#A sequence of x values and y values for plotting the trendline
x_fit = np.linspace(x_new_data.min(), x_new_data.max(), 50)
y_fit = polynomial(x_fit)

#Plotting
plt.plot(x_fit, y_fit, color='blue', label=f'Trendline: {coefficients[0]:.3E}*x + {coefficients[1]:.3E}')

#Scatter Plot
plt.scatter(x_data, y_data, color='blue', label='Delay')
plt.scatter(x_new_data, y_new_data, color='red', marker='^', label='Measrued')

# Add labels and a title
plt.xlabel('distance (m)')
plt.ylabel('delay (ns)')
plt.title('Delay vs. Distance (1 Mhz)')
plt.legend()
plt.show()

```
